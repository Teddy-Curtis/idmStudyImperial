```python
import json 
import matplotlib.pyplot as plt
import numpy as np 
```

```python
procs = ["h2h2lPlM", "h2h2lPlMnunu", "both"]
xses = {}
for proc in procs:
    with open(f"plots/{proc}/xses.json", 'r') as f:
        xs = json.load(f)
    diff = []
    for i in range(len(xs)-1):
        diff.append(xs[i] / xs[i+1])
    xses[proc] = [xs, diff]

mH = [60, 100, 140, 180, 220]
print(diff)
fig, axs = plt.subplots(2)
for proc, values in xses.items():
    axs[0].scatter(mH, values[0], label=proc)
    axs[1].scatter([1, 2, 3, 4], values[1], label=proc)

axs[0].set_ylabel("Cross-section")
axs[0].set_xlabel("mH")
axs[0].legend()
axs[0].set_yscale("log")

axs[1].legend()
axs[1].set_yscale('log')

plt.tight_layout()
```

```python

with open(f"mAmH30_mHchmH40/plots/h2h2lPlM/xses.json", 'r') as f:
    xs_30 = json.load(f)
    
with open(f"mAmH10_mHchmH30/plots/h2h2lPlM/xses.json", 'r') as f:
    xs_10 = json.load(f)

xs_30 = np.array(xs_30)
xs_10 = np.array(xs_10)

# normalise 
xs_30 = xs_30 / xs_30[0]
xs_10 = xs_10 / xs_10[0]

mHs = [80, 100, 140, 180, 220]

plt.scatter(mHs, xs_30, label="mA-mH=30, mHch-mH=40", alpha=0.8)
plt.scatter(mHs, xs_10, label="mA-mH=10, mHch-mH=30", alpha=0.8)
plt.legend()
plt.xlabel("mH (GeV)")
plt.ylabel("Normalised xs")
plt.yscale("log")
plt.title("xs vs mH for given mass splits")
```
